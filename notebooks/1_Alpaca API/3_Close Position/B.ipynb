# Close position

## Prepare client

## API Key

```python
import os

API_KEY = os.getenv('API_KEY')
API_SECRET = os.getenv('API_SECRET')
```

## Instantiate Python client for Alpaca API

```python
from alpaca.trading.client import TradingClient
client = TradingClient(API_KEY, API_SECRET, paper=True)
```

## Account status

### Orders

```python
import pandas as pd
from alpaca.trading import requests

request = requests.GetOrdersRequest(status='all')
orders = client.get_orders(request)

pd.DataFrame([order.dict() for order in orders])
```

### Positions

```python
positions = client.get_all_positions()

df = pd.DataFrame([position.dict() for position in positions]).set_index('symbol')
df
```

## Closing position

### Select symbol

```python
symbol = 'BTCUSD'
```

### Prepare request

```python
request = requests.ClosePositionRequest(percentage='10')
```

### Submit the request to close position

```python
response = client.close_position(symbol_or_asset_id=symbol, close_options=request)
```

```python
pd.DataFrame(response.dict(), index=[0])
```
